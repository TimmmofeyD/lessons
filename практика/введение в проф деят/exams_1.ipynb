```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exams = pd.read_csv("./exams.csv")
exams.info()
```

```python
exams.head()
```

```python
exams['parental level of education'].value_counts()[:10].plot(kind = 'pie' , autopct = '%1.1f%%' , shadow = True , explode = [0,0,0,0,0.1,0])
fig = plt.gcf()
fig.set_size_inches(7,7)
plt.show()
```

```python
sns.countplot(y=exams['gender']);
```

```python
fig, ax = plt.subplots(1,3)
sns.barplot(x=exams['gender'],y=exams['reading score'], ax=ax[0])
sns.barplot(x=exams['gender'],y=exams['writing score'],ax=ax[1])
sns.barplot(x=exams['gender'],y=exams['math score'],ax=ax[2]);
```
