```python
counts = Counter(np.around(data["Open"]))
x_1 = []
y_1 = []
for item, count in counts.items():
    x_1.append(count)
    y_1.append(item)
y = x_1
x = y_1
plt.figure(figsize=(40, 20))
plt.title(' ', size = 55)
plt.xlabel('', size = 55)
plt.ylabel('', size = 55)
plt.tick_params(labelsize=35)
plt.plot(x, y, marker = 'o', linewidth = 4)
```

```python
plt.figure(figsize=(40, 20))

plt.title(' $  ', size = 55)

plt.xlabel(' $', size = 55)
plt.rcParams.update({'font.size': 28})

plt.hist(data['Open'], 20, edgecolor = (0, 0, 0),density=True, label="Data", linewidth = 2.5, rwidth=0.9)
plt.xticks([10 * i for i in range(16)])

mu = 65
variance = 55
sigma = math.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 200)
plt.plot(x, stats.norm.pdf(x, mu, sigma), linewidth = 8);
```

```python
def shift(n, x, tau):
    return (1 / (n - tau)) * sum([abs((t + tau) - t) for t in x])
```

```python
count = data["Open"].count()
count
```

```python
plt.figure(figsize=(40, 20))

y = shift(count, data['Open'], 0.01) 
x = [i for i in data["Open"]]

plt.plot(x, y, linewidth = 4, marker = "o")
```
