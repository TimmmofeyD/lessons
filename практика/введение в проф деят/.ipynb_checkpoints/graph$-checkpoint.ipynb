```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker

plt.style.use('seaborn-whitegrid')
```

```python
data = pd.read_csv('./usd_to_rub.csv', sep=',')
```

```python
data.head()
```

```python
data = data.drop('Volume', axis = 1)
```

```python
data['Adj Close'].hist();
```

```python
yearmoth = data["Date"].apply(lambda x: x[0:7])
```

```python
y = data['Adj Close']
x = yearmoth
plt.figure(figsize=(50, 20))
plt.title(' $    ', size = 55)
plt.ylabel(' $', size = 55)
plt.xlabel('', size = 55)
plt.xticks(rotation=90)
plt.tick_params(labelsize=25)
plt.plot(x, y)
```
